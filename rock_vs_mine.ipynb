```python
# Install LightGBM if not installed (only needed once)
# !pip install lightgbm

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import lightgbm as lgb

# Load your dataset (uploaded file)
sonar_data = pd.read_csv('/content/sonar data.csv', header=None)

# Check data
print(sonar_data.head())
print(sonar_data.shape)
print(sonar_data[60].value_counts())

# Separate features and target
X = sonar_data.drop(columns=60, axis=1)
Y = sonar_data[60].astype(str)  # Ensure target is string for classification

# Split data into train and test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, stratify=Y, random_state=1)

# Initialize LightGBM classifier
model = lgb.LGBMClassifier()

# Train the model
model.fit(X_train, Y_train)

# Predict on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print('Accuracy on training data:', training_data_accuracy)

# Predict on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print('Accuracy on test data:', test_data_accuracy)

# Making a prediction on a new input
input_data = (0.0762,0.0666,0.0481,0.0394,0.0590,0.0649,0.1209,0.2467,0.3564,0.4459,0.4152,0.3952,
              0.4256,0.4135,0.4528,0.5326,0.7306,0.6193,0.2032,0.4636,0.4148,0.4292,0.5730,0.5399,
              0.3161,0.2285,0.6995,1.0000,0.7262,0.4724,0.5103,0.5459,0.2881,0.0981,0.1951,0.4181,
              0.4604,0.3217,0.2828,0.2430,0.1979,0.2444,0.1847,0.0841,0.0692,0.0528,0.0357,0.0085,
              0.0230,0.0046,0.0156,0.0031,0.0054,0.0105,0.0110,0.0015,0.0072,0.0048,0.0107,0.0094)

# Convert input data to numpy array
input_as_numpy_array = np.asarray(input_data)

# Reshape the numpy array for prediction
input_data_reshaped = input_as_numpy_array.reshape(1, -1)

# Make prediction
prediction = model.predict(input_data_reshaped)
print('Prediction:', prediction[0])

if prediction[0] == 'R':
    print('The object is a Rock')
else:
    print('The object is a Mine')
```
